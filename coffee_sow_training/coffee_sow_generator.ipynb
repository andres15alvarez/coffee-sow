```python
import numpy as np 
import pandas as pd
```

```python
def sow(temperature, humidity, altitude, rain, sunshine):
  """
  Args:
    temperature(float): temperature measure in degrees centigrades.
    humidity(float): relative humidity in the air.
    altitude(float): masl of the place to sow.
    rain(float): milimeters of rain in a squared meter in a month.
    sunshine(float): hours of sun in a day.

  Output:
  sow(int): 1 if is good to sow, 0 otherwise.
  """
  sow = None
  if (19 <= temperature <= 22 and 0.5 <= humidity <= 0.8 and 800 <= altitude <= 2100 and 120 <= rain <= 180 and 4.5 <= sunshine <= 5.5):
    sow = 1
  else:
    sow = 0
  return sow
```

```python
temperature = np.round(np.random.normal(20, 2.5, 10000), 2)
humidity = np.round(np.random.normal(0.6, 0.08, 10000), 4)
altitude = np.round(np.random.normal(1500, 100, 10000), 2)
rain = np.round(np.random.normal(150, 10, 10000), 2)
sunshine = np.round(np.random.normal(5.0, 0.1, 10000), 2)
```

```python
to_sow = np.array([sow(temperature[i], humidity[i], altitude[i], rain[i], sunshine[i]) for i in range(10000)])
dictionary = {
    'temperature': temperature, 
    'humidity': humidity, 
    'altitude': altitude, 
    'rain': rain, 
    'sunshine': sunshine, 
    'sow': to_sow
    }
df = pd.DataFrame(dictionary)
```
